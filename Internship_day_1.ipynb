NumPy (Numerical Python) ACTIVITY DAY 3

```python
import numpy as np
a=np.array([10,20,30,40])
b=np.array([5,10,15,20])
add = a+b
print("Addition:",add)
sub = a-b
print("Subtraction:",sub)
mul = a*b
print("Multiplication:",mul)
div = a/b
print("Division:",div)
total = np.sum(a)
print("Sum of a:",total)
mean = np.mean(a)
print("Mean of a:",mean)
```
